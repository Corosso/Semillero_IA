#**Cargar Data**

```python
#Cargamos df con informacion de las 31 grabadoras
df_grabadoras = pd.read_csv('/content/drive/MyDrive/Bio_Acustica/df_grabadoras.csv')

#cargamos los audios separados segun su cluster
with open('/content/drive/MyDrive/Bio_Acustica/data.pkl', 'rb') as f:
    pk_clusters = pickle.load(f)
```

#Clusterizando

```python
df_grabadoras_cluster1=df_grabadoras.copy()
```

```python
df_grabadoras_cluster1["Cluster 1"]=0
df_grabadoras_cluster1["Cluster 2"]=0
df_grabadoras_cluster1["Cluster 3"]=0

num_clusters = np.shape(pk_clusters)[0]
for i in range(num_clusters):
    num_audios_cluster = np.shape(pk_clusters[i])[0]
    for j in range(num_audios_cluster):
        nom_aud=os.path.basename(pk_clusters[i][j])
        nom_grab=nom_aud.split("_")[0]
        df_grabadoras_cluster1.loc[df_grabadoras_cluster1["Recorder"] == nom_grab, "Cluster "+str(i+1)] += 1

```

## Distribucion en mapa

```python
# Crear un diccionario para asignar colores a cada cluster
colores_clusters = {
    "Cluster 1": 'blue',
    "Cluster 2": 'green',
    "Cluster 3": 'red',
}
# Coordenadas del punto

lat = 6.4137
lng = -75.0366

# Crear mapa centrado en las coordenadas
m = folium.Map(location=[lat, lng], zoom_start=12)

# Agregar círculo azul al mapa
for i in df_grabadoras_cluster1.index:
  #Ubicacion grabadora
  lat,lng=df_grabadoras_cluster1["Latitud"][i],df_grabadoras_cluster1["Longitud"][i]
  if df_grabadoras_cluster1["Cluster 1"][i]>0:
    tooltip = df_grabadoras_cluster1.columns[8]
    radio= int(df_grabadoras_cluster1["Cluster 1"][i]/6000*18000)
    folium.Circle(location=[lat, lng], radius=radio, color=colores_clusters["Cluster 1"], fill=True, fill_opacity=0.3, tooltip=tooltip).add_to(m)
  if df_grabadoras_cluster1["Cluster 2"][i]>0:
    tooltip = df_grabadoras_cluster1.columns[9]
    radio= int(df_grabadoras_cluster1["Cluster 2"][i]/6000*18000)
    folium.Circle(location=[lat, lng], radius=radio, color=colores_clusters["Cluster 2"], fill=True, fill_opacity=0.3, tooltip=tooltip).add_to(m)
  if df_grabadoras_cluster1["Cluster 3"][i]>0:
    tooltip = df_grabadoras_cluster1.columns[10]
    radio= int(df_grabadoras_cluster1["Cluster 3"][i]/6000*18000)
    folium.Circle(location=[lat, lng], radius=radio, color=colores_clusters["Cluster 3"], fill=True, fill_opacity=0.3, tooltip=tooltip).add_to(m)
  icon = icon = folium.Icon(icon='glyphicon-map-marker', prefix='glyphicon', icon_color='black', icon_size=(9, 9))
  popup_text=df_grabadoras_cluster1["Recorder"][i],"Lat:",df_grabadoras_cluster1["Latitud"][i],"Lng:",df_grabadoras_cluster1["Longitud"][i],"Elev:",df_grabadoras_cluster1["Elevacion"][i] ,"Cluster1:",df_grabadoras_cluster1["Cluster 1"][i],"Cluster2:",df_grabadoras_cluster1["Cluster 2"][i],"Cluster3:",df_grabadoras_cluster1["Cluster 3"][i],df_grabadoras_cluster1["Habitat"][i],"Grabaciones:",df_grabadoras_cluster1["Total_grabaciones"][i]
  folium.Marker(location=[lat, lng], popup=popup_text, icon=icon, opacity=0.3).add_to(m)
m.save("/content/drive/MyDrive/Bio_Acustica/mapa_grabadoras_cluster.html")
m
```
